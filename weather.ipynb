```python
from mpl_toolkits.basemap import Basemap,cm
import numpy as np
import matplotlib.pyplot as plt
import netCDF4 as nc
%matplotlib inline
```

```python
fileName = r'C:\Users\lenovo\Desktop\Map\China.nc'
data = nc.Dataset(fileName)
```

```python
data
```

```python
lat = data.variables['LAT'][:]
lon = data.variables['LON'][:]
t = data.variables['TAIR_MEAN'][:]
```

```python
m = Basemap(projection = 'cyl',llcrnrlat = min(lat)-0.5,urcrnrlat = max(lat) +0.5,
            llcrnrlon = min(lon) -0.5,urcrnrlon = max(lon) +0.5,resolution = 'c')
m.drawcoastlines()
m.drawstates()
m.drawcountries()

m.drawparallels(np.arange(-90,91,30))
m.drawmeridians(np.arange(-180,181,60))
m.drawmapboundary(fill_color = [0.9,0.9,1])

x, y = m(*np.meshgrid(lon,lat))
#clevs = [-20,-16,-12,-8,-4,0,4,8,12,16,20]
cs = m.contourf(x,y,t)
cbar = m.colorbar(cs,location = 'bottom',pad='5%')
cbar.set_label('$^oC$')
plt.show()
```
